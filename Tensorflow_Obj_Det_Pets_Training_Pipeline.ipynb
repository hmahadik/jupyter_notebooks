````
! cd /tensorflow/models/research && \
  export PYTHONPATH=$PYTHONPATH:`pwd`:`pwd`/slim && \
  wget -nc http://www.robots.ox.ac.uk/~vgg/data/pets/data/images.tar.gz  && \
  wget -nc http://www.robots.ox.ac.uk/~vgg/data/pets/data/annotations.tar.gz  && \
  tar -xf annotations.tar.gz  && \
  tar -xf images.tar.gz  && \
  python object_detection/dataset_tools/create_pet_tf_record.py \
      --label_map_path=object_detection/data/pet_label_map.pbtxt \
      --data_dir=`pwd` \
      --output_dir=`pwd`
````

Recommended Structure for Training & Eval

```
+data
  -label_map file
  -train TFRecord file
  -eval TFRecord file
+models
  + model
    -pipeline config file
    +train
    +eval
```



# Train & Evaluate

````
! cd /tensorflow/models/research && \
    mkdir -p /running_pets && \
    mkdir -p /running_pets/data && \
    mkdir -p /running_pets/models && \
    mkdir -p /running_pets/models/model && \
    cp -v pet_faces_train.record-* /running_pets/data && \
    cp -v pet_faces_val.record-* /running_pets/data && \
    cp -v object_detection/data/pet_label_map.pbtxt /running_pets/data && \
    wget -nc http://download.tensorflow.org/models/object_detection/ssd_mobilenet_v1_coco_2018_01_28.tar.gz && \
    tar -xf ssd_mobilenet_v1_coco_2018_01_28.tar.gz && \
    cp -v ssd_mobilenet_v1_coco_2018_01_28/model.ckpt.* /running_pets/data/ && \
    cat object_detection/samples/configs/ssd_mobilenet_v1_pets.config && \
    cp -v object_detection/samples/configs/ssd_mobilenet_v1_pets.config \
        /running_pets/data/ssd_mobilenet_v1_pets.config && \
    sed -i "s|PATH_TO_BE_CONFIGURED|/running_pets/data|g" \
        /running_pets/data/ssd_mobilenet_v1_pets.config && \
    ls -lah /running_pets/* && \
    cat /running_pets/data/ssd_mobilenet_v1_pets.config
````

Before running the training, evaluate the pre-trained network as-is.
Then, train and evaluate as you train.

````

! cd /tensorflow/models/research/ && \
    export PYTHONPATH=$PYTHONPATH:`pwd`:`pwd`/slim && \
    PIPELINE_CONFIG_PATH=/running_pets/data/ssd_mobilenet_v1_pets.config && \
    MODEL_DIR=/running_pets/models/model && \
    NUM_TRAIN_STEPS=200000 && \
    SAMPLE_1_OF_N_EVAL_EXAMPLES=1 && \
    rm -rfv ${MODEL_DIR} && \
    mkdir -p ${MODEL_DIR} && \
    echo "Training" && \
    echo "--------" && \
    python object_detection/model_main.py \
        --pipeline_config_path=${PIPELINE_CONFIG_PATH} \
        --model_dir=${MODEL_DIR} \
        --num_train_steps=${NUM_TRAIN_STEPS} \
        --sample_1_of_n_eval_examples=$SAMPLE_1_OF_N_EVAL_EXAMPLES \
        --alsologtostderr
````

````
!ls -lah /running_pets/models/*
````

````
! cd /tensorflow/models/research/ && \
    export PYTHONPATH=$PYTHONPATH:`pwd`:`pwd`/slim && \
    PIPELINE_CONFIG_PATH=/running_pets/data/ssd_mobilenet_v1_pets.config && \
    MODEL_DIR=/running_pets/models/model && \
    NUM_TRAIN_STEPS=50000 && \
    SAMPLE_1_OF_N_EVAL_EXAMPLES=1 && \
    echo "Evaluating (pre-trained)" && \
    echo "------------------------" && \
    python object_detection/model_main.py \
        --pipeline_config_path=${PIPELINE_CONFIG_PATH} \
        --model_dir=${MODEL_DIR} \
        --run_once \
        --alsologtostderr
````

# Export Frozen Model

````
!   rm -rf /running_pets/models/model && \
    mkdir -p /running_pets/models/model && \
    cd /tensorflow/models/research/ && \
    export PYTHONPATH=$PYTHONPATH:`pwd`:`pwd`/slim && \
    INPUT_TYPE=image_tensor && \
    PIPELINE_CONFIG_PATH=/running_pets/data/ssd_mobilenet_v1_pets.config && \
    TRAINED_CKPT_PREFIX=/running_pets/data/model.ckpt && \
    EXPORT_DIR=/running_pets/models/model && \
    python object_detection/export_inference_graph.py \
        --input_type=${INPUT_TYPE} \
        --pipeline_config_path=${PIPELINE_CONFIG_PATH} \
        --trained_checkpoint_prefix=${TRAINED_CKPT_PREFIX} \
        --output_directory=${EXPORT_DIR}
````
